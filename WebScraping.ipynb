```
rating=[]
review=[]
for page in range(2,200):
  wd.get('https://www.flipkart.com/redmi-6-black-64-gb/product-reviews/itmfa94eqhsxwhyn?pid=MOBFA2NXDQFTXNHJ&page=' + str(page))
  dat=wd.find_element_by_class_name('_1HmYoV')
  for i in range(3,13):
    record=dat.find_elements_by_xpath("//*[@id='container']/div/div[3]/div/div/div[2]/div[" + str(i) + "]")
    a=dat.find_elements_by_xpath("//*[@id='container']/div/div[3]/div/div/div[2]/div["+ str(i) + "]/div/div/div/div[1]/div")
    b=dat.find_elements_by_xpath("//*[@id='container']/div/div[3]/div/div/div[2]/div["+ str(i) + "]/div/div/div/div[2]/div")
    for a_rating in a:
      rating.append(a_rating.text)
    for b_review in b:
      review.append(b_review.text)
table = pd.DataFrame(review,columns=['Review'])
table ["Rating"] = rating
table
```

```
table.to_csv("Redmi6Black64GB.csv")
```
